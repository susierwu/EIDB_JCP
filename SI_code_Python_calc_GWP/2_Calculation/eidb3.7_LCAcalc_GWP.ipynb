```python
GWP = [method for method in bw.methods if "climate change" in str(method) and "IPCC 2013" in str(method) and 'GWP 100a' in str(method) and 'no LT' not in str(method)]
```

### final calc; use all len(df_lca_calc) in for loop

```python
x = [[]]*len(df_lca_calc)
final_df = pd.DataFrame()

#only need to fulfill two conditions in "all_prod_list", will incl. all relevant products, do not incl prod.name = act['name'] as one uniq ref.prod. has many act, the read-in Excel file only saved one of them in prod_name col.
for i in range(0,len(df_lca_calc)):
    a = qlca.LCAinput_manual()
    a.all_prod_list = [act for act in eidb if act['reference product'] == df_lca_calc.ref_prod[i] and "ordinary transforming activity" in act['activity type']]
    #a.showLCI() this line will print all prod. name for each ref. prod
    x[i] = qlca.LCAcalc(a.prodFU(), a.prodname(), GWP)    
    isicvalue = list(df_lca_calc.loc[df_lca_calc['ref_prod'] == df_lca_calc.ref_prod[i],'ISIC_cat']) * list(df_lca_calc.loc[df_lca_calc['ref_prod'] == df_lca_calc.ref_prod[i],'prod_freq'])[0]          
    y = pd.DataFrame()
    y = x[i].T
    try:
        y["ISIC"] = isicvalue
    except ValueError:
        y["ISIC"] = "various/NA ISIC"
    final_df = final_df.append(y)
```

### create dataframe for each uniq ref. product, built from the calculated x[i] dataframe:

```python
uniq_final_df = pd.DataFrame()
for i in range(0,len(x)):
    try:
        pCOV = stat.stdev(list(x[i].iloc[0])) / stat.mean(list(x[i].iloc[0]))
    except ZeroDivisionError:
        pCOV = 0
    pname = df_lca_calc.ref_prod[i] #x[i].T.index[0]  #unique ref prod. name
    pCat = df_lca_calc.ISIC_cat[i]                    
    prod_n = list(df_lca_calc.loc[df_lca_calc['ref_prod'] == df_lca_calc.ref_prod[i],'prod_freq'])[0] #total num of diff. production
    rdt = {'ISIC':[pCat], 'uniq_RefProd':[pname], '#diff_prod':[prod_n], 'COV': [pCOV]}
    rr = pd.DataFrame(rdt)
    uniq_final_df = uniq_final_df.append(rr)
```
